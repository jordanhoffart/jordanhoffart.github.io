```python
import sympy as sp
from sympy.plotting import plot
```

```python
# 1a
n = sp.symbols('n')
bn = 1 / (12 * n + 100)
upper_bound = bn.subs(n,n+1)
tolerance = 0.001
plot(upper_bound, (n,1,100), tolerance)
```

```python
# 1b
eq = sp.Eq(upper_bound, tolerance)
N = sp.nsolve(eq, n, 80)
N = int(N)
print(N)
```

```python
# 1c
an = (-1)**n * bn
sum = sp.summation(an, (n,1,N))
sum = sum.evalf()
print(sum)
```

```python
# 1d
absolute_sum = sp.summation(bn, (n,1,sp.oo))
print(absolute_sum)
print('series is not absolutely convergent')
```

```python
# 2a
for i in range(16):
    ai = sp.factorial(i) / 5000**i
    ai = ai.evalf()
    print(ai)
```

```python
#2b
an = sp.factorial(n) / 5000**n
ratio = an.subs(n,n+1) / an
limit = sp.limit(ratio, n, sp.oo)
print(limit)
```

```python
#2c
print('series diverges')
```

```python
#3a
x = sp.symbols('x')
bn = x**(8*n+4) / (6*n+3)
ratio = bn.subs(n,n+1) / bn
limit = sp.limit(ratio, n, sp.oo)
print(limit)
```

```python
#3b
an = (-1)**n * bn
an_left = an.subs(x,-1)
an_right = an.subs(x,1)
left = sp.summation(an_left, (n,0,sp.oo))
right = sp.summation(an_right, (n,0,sp.oo))
print(left)
print(right)
print('interval of convergence: [-1,1]')
```

```python
#3c
s1 = sp.summation(an, (n,0,1))
s3 = sp.summation(an, (n,0,3))
s5 = sp.summation(an, (n,0,5))
print(1,s1)
print(3,s3)
print(5,s5)
```

```python
#3d
f = sp.atan(x**4) / 3
plot(s1,s3,s5,f,(x,-1,1))
```
