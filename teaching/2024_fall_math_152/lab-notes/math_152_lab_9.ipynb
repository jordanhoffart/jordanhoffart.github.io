```python
import sympy as sp
from sympy.plotting import plot
```

```python
print("problem 1a")
x = sp.symbols('x', real=True)
n = sp.symbols('n', positive=True, integer=True)
numerator = sp.factorial(2*n)*x**(2*n+1)
denominator = 2**(2*n)*sp.factorial(n)**2*(2*n+1)
summand = numerator / denominator
ratio = sp.Abs(summand.subs(n,n+1) / summand)
limit = sp.limit(ratio, n, sp.oo)
display(limit)
print("radius of convergence is 1")
```

```python
print("problem 1b")
partial_sums = [sp.summation(summand,(n,0,N)) for N in [1,5,10]]
plot(*partial_sums, sp.asin(x), (x, -1, 1))
```

```python
print("problem 2a")
def taylor_polynomial(f, a, degree):
    taylor_polynomial = 0
    for n in range(degree + 1):
        coefficient = sp.diff(f,x,n).subs(x,a) / sp.factorial(n)
        taylor_polynomial += coefficient * (x-a)**n
    return taylor_polynomial

degree = 10
f = sp.cos(x)
a = 0
display(taylor_polynomial(f,a,degree))
```

```python
print("problem 2b")
f = sp.ln(x)
a = sp.E
display(taylor_polynomial(f,a,degree))
```

```python
print("problem 2c")
f = sp.csc(x)
a = sp.pi/2
display(taylor_polynomial(f,a,degree))
```
